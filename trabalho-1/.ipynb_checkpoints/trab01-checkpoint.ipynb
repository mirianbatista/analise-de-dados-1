```r
titanic <- read.csv("titanic.csv")
titanic
```

```r
set.seed(12)
```

```r
random <- sample(1:nrow(breastcancer), 0.7 * nrow(breastcancer))
randomiris <- sample(1:nrow(iris), 0.7 * nrow(iris))
randomtitanic <- sample(1:nrow(titanic), 0.7 * nrow(titanic))
```

```r
breastcancer_train <- optbin(breastcancer[random, ], method = "infogain")
iris_train <- optbin(iris[randomiris, ], method = "infogain")
titanic_train <- optbin(titanic[randomtitanic, ], method = "infogain")
```

```r
breastcancer_test <- breastcancer[random, ]
iris_test <- iris[randomiris, ]
titanic_test <- titanic[randomtitanic, ]
```

```r
model_train_breastcancer <- OneR(breastcancer_train, verbose = TRUE)
```

```r
model_train_iris <- OneR(iris_train, verbose = TRUE)
```

```r
model_train_titanic <- OneR(titanic_train, verbose = TRUE)
```

```r
summary(model_train_breastcancer)
```

```r
summary(model_train_iris)
```

```r
summary(model_train_titanic)
```

```r
naivetrain_cancer <- breastcancer[random, ]
naivetrain_iris <- iris[randomiris, ]
naivetrain_titanic <- titanic[randomtitanic, ]
```

```r
naivetest_cancer <- breastcancer[-random, ]
naivetest_iris <- iris[-randomiris, ]
naivetest_titanic <- titanic[-randomtitanic, ]
```

```r
breast_naive <- naiveBayes(Class ~ ., data=naivetrain_cancer)
iris_naive <- naiveBayes(Species ~ ., data=naivetrain_iris)
titanic_naive <- naiveBayes(survived ~ ., data=naivetrain_titanic)
```

```r
prednaive_cancer <- predict(breast_naive, naivetest_cancer[, -10])
prednaive_iris <- predict(iris_naive, naivetest_iris[, -10])
prednaive_titanic <- predict(titanic_naive, naivetest_titanic[, -10])
```

```r
table(prednaive_cancer, naivetest_cancer[, 10])
```

```r
table(prednaive_iris, naivetest_iris[, 10])
```

```r
table(prednaive_titanic, naivetest_titanic[, 10])
```
