```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
alunos = pd.read_csv('alunosrt.csv')
alunos
```

```python
#peso por curso
#get_group n tá pegando
peso_curso = alunos['peso'].groupby(alunos['curso'])
peso_curso
```

```python
peso_curso.value_counts()
```

```python
peso_design = alunos.groupby('peso').curso.agg(lambda g: g.eq('Design').count())
peso_design
```

```python
pd = peso_design.sort_values(ascending=False)
pd
```

```python
pd.plot(kind='bar')
plt.title("Peso dos alunos de Design")
plt.ylabel('Quantidade de Alunos')
plt.xlabel('Peso dos Alunos')
```
